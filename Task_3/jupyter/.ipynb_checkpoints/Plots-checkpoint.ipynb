```python
import numpy as np
import matplotlib.pyplot as plt

N = 100  # количество узлов по xi
M = 100  # количество узлов по eta
L_xi = 1.0  # длина по xi
L_eta = 2 * np.pi  # длина по eta
h_xi = L_xi / N
h_eta = L_eta / M

# Параметры задачи
gamma = 1.4  # показатель адиабаты
M_inf = 0  # число Маха на бесконечности

xi = np.linspace(0, L_xi, N + 1)
eta = np.linspace(0, L_eta, M + 1)
Phi = np.zeros((N + 1, M + 1))
rho = np.ones((N + 1, M + 1))

# Начальные условия
for i in range(N + 1):
    for j in range(M + 1):
        Phi[i, j] = np.exp(xi[i]) * np.sin(eta[j])

# Граничные условия
Phi[0, :] = 0  # Phi_{0,j} = 0, xi = 0
Phi[-1, :] = np.exp(xi[-1]) * np.sin(eta)  # Phi_{N, j} = e^{\xi_N} \sin \eta_j, xi = xi_inf
Phi[:, 0] = 0  # Phi_{i, 0} = 0, eta = 0
Phi[:, -1] = 0  # Phi_{i, M} = 0, eta = 2*pi
```

```python
def Plot(xi, eta, Phi):
    fig = plt.figure(figsize=(10, 10))
    R, Theta = np.meshgrid(xi, eta)
    ax = fig.add_subplot(111, projection='polar')
    c = ax.pcolormesh(Theta, R, Phi.T, cmap='viridis')
    plt.colorbar(c, ax=ax, label='Phi')

    plt.show()

# Использование функции
Plot(xi, eta, Phi)
```

```python
num_steps = 1

# Явный метод
for step in range(num_steps):
    for i in range(1, N):
        for j in range(1, M):
            # Расчет q_ij только для внутренних узлов
            q_sq = np.exp(2 * xi[i]) * (
                    ((Phi[i + 1, j] - Phi[i - 1, j]) / (2 * h_xi)) ** 2 +
                    ((Phi[i, j + 1] - Phi[i, j - 1]) / (2 * h_eta)) ** 2
            )
            
            # Расчет rho_ij
            rho[i, j] = ((gamma - 1) / 2 * M_inf ** 2 * (1 - q_sq)) ** (1 / (gamma - 1))

            # Обновление Phi только для внутренних узлов
            Phi[i, j] = (
                    rho[i + 1, j] * Phi[i + 1, j] +
                    rho[i - 1, j] * Phi[i - 1, j] +
                    rho[i, j + 1] * Phi[i, j + 1] +
                    rho[i, j - 1] * Phi[i, j - 1]
            ) / (rho[i + 1, j] + rho[i - 1, j] + rho[i, j + 1] + rho[i, j - 1])
```

```python
num_steps = 10000

# Явный метод
for step in tqdm(range(num_steps)):
    for i in range(1, N):
        for j in range(1, M):
            # Обновление Phi только для внутренних узлов
            Phi[i, j] = (
                    rho[i + 1, j] * Phi[i + 1, j] +
                    rho[i - 1, j] * Phi[i - 1, j] +
                    rho[i, j + 1] * Phi[i, j + 1] +
                    rho[i, j - 1] * Phi[i, j - 1]
            ) / (rho[i + 1, j] + rho[i - 1, j] + rho[i, j + 1] + rho[i, j - 1])
```

```python
Plot(xi, eta, Phi)
```
