```python
import matplotlib.pyplot as plt
import numpy as np

def integrand(theta, z, rq):
    return (1 - rq * np.cos(theta)) / np.sqrt(rq**2 + 1 - 2 * rq * np.cos(theta)) * z / np.sqrt(z**2 - (rq**2 + 1 - 2 * rq * np.cos(theta)))

def rectangular_integration(z, rq, n):
    integral_sum = 0.0
    for k in range(1, n + 1):
        theta_k = 2 * np.pi * k / n
        integral_sum += integrand(theta_k, z, rq)
    integral = rq * 2 * np.pi / n * integral_sum
    return integral

z_values = np.linspace(0, 5, 100)
rq_value = 2
n_value = 1000

F_values = [rectangular_integration(z_val, rq_value, n_value) for z_val in z_values]

plt.figure(figsize=(8, 6))
plt.plot(z_values, F_values, label=f'rq={rq_value}')
plt.xlabel('z')
plt.ylabel('F(z)')
plt.legend()
plt.grid(True)
plt.show()
```
